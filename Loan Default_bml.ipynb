```python
import pandas as pd
# Handle missing values
# Example: Impute missing values in numeric columns with mean, and fill missing values in categorical columns with a placeholder
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

df = df.fillna({
    'CreditGrade': 'Unknown',
    'Occupation': 'Unknown',
    'EmploymentStatus': 'Unknown',
    'ClosedDate': 'Unknown',
    'GroupKey': 'Unknown',
    # Fill other columns as needed
})

# Remove unnecessary columns
# Example: Drop columns with too many missing values or columns that are not relevant for analysis
columns_to_drop = [
    'ListingKey', 'ListingNumber', 'FirstRecordedCreditLine', 'DateCreditPulled',
    'LoanKey', 'MemberKey', 'LoanOriginationDate', 'LoanOriginationQuarter',
    # Add other columns to drop as needed
]
df = df.drop(columns_to_drop, axis=1)

df.head()
```

```python
# Define the mapping dictionary for the new labels
mapping = {
    'Current': 'Non-Default',
    'Completed': 'Non-Default',
    'Chargedoff': 'Default',
    'Defaulted': 'Default',
    'Past Due (1-15 days)': 'Non-Default',
    'Past Due (31-60 days)': 'Non-Default',
    'Past Due (61-90 days)': 'Non-Default',
    'Past Due (91-120 days)': 'Non-Default',
    'Past Due (16-30 days)': 'Non-Default',
    'FinalPaymentInProgress': 'Non-Default',
    'Past Due (>120 days)': 'Non-Default'
}

# Create a new column with the binary labels
df['LoanStatus_Binary'] = df['LoanStatus'].map(mapping)
```
